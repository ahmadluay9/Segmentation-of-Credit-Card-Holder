# Model Inference

> This section explains the process of saving and loading models and other files related to the results of the models that we created in the previous chapter. Since there are only 10 data, handling outliers could be skipped.

```python
# Importing Libraries
import pandas as pd

# Save and Load Model
import pickle
import json
```

```python
# Load Model Files

with open('kmeans.pkl', 'rb') as file_1:
  kmeans = pickle.load(file_1)

with open('s_minmaxscaler.pkl', 'rb') as file_2:
  s_minmaxscaler = pickle.load(file_2)

with open('pca.pkl', 'rb') as file_3:
  pca = pickle.load(file_3)
```

## Load Data Inference

```python
# Load Data Inference
df_inf = pd.read_csv('inference_Credit_Card_Customer_Segmentation.csv')
df_inf = df_inf.drop(['Unnamed: 0'],axis=1)
df_inf = df_inf.set_index('CUST_ID')
df_inf_copy = df_inf.copy()
df_inf_copy = df_inf.drop(['TENURE'],axis=1)
df_inf_copy
```

## Feature Scaling

```python
# Feature Scaling for data inference
df_inf_scaled = s_minmaxscaler.transform(df_inf_copy)
df_inf_scaled = pd.DataFrame(df_inf_scaled, columns = df_inf_copy.columns, index = df_inf_copy.index)
df_inf_scaled
```

##  Model Training Inference

```python
# PCA analysis
pca_inf = pca.transform(df_inf_scaled)
pca_inf
```

```python
# Cluster inference
y_inf = kmeans.predict(pca_inf)

# Create DataFrame
y_inf = pd.DataFrame(y_inf, columns=['Customer Segmentation (Inference)'],index=df_inf_scaled.index)
y_inf
```

```python
# Concatinate with Data inference
model_inf = pd.concat([df_inf, y_inf['Customer Segmentation (Inference)']], axis=1).sort_index()
model_inf
```

Based on the table above, we can see the Holders/Customers Segmentation from data inference shows that there are:

- 2 customers segmented as Cluster 0 (credit card holders who frequently make purchases)
- 6 customers segmented as Cluster 1 (credit card holders who don't use their credit cards very frequently to make purchases)
- 2 customers segmented as Cluster 2 (credit card holders who frequently use Cash in advance payment method when making purchases)
